```python
df_all = pd.merge(df,df_other,on='FILM',how='inner')
```

```python
df_all.info()
```

```python
df_all
```

```python
df_all['RT_Norm'] = np.round(df_all['RottenTomatoes']/20,1)
```

```python
df_all['RT_user_Norm'] = np.round(df_all['RottenTomatoes_User']/20,1)
```

```python
df_all['Met_Norm'] = np.round(df_all['Metacritic']/20,1)
```

```python
df_all['Met_user_Norm'] = np.round(df_all['Metacritic_User']/2,1)
```

```python
df_all['IMDB_Norm'] = np.round(df_all['IMDB']/2,1)
```

```python
df_all.head()
```

```python
df_norm = df_all[['STARS','RATING','RT_Norm','RT_user_Norm','Met_Norm','Met_user_Norm','IMDB_Norm']]
```

```python
df_norm.head()
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.kdeplot(data=df_norm,x='STARS',fill=True,clip=(0,5),label='Stars')
sns.kdeplot(data=df_norm,x='RATING',fill=True,clip=(0,5),label='rating')
sns.kdeplot(data=df_norm,x='RT_Norm',fill=True,clip=(0,5),label='RT_Norm')
sns.kdeplot(data=df_norm,x='RT_user_Norm',fill=True,clip=(0,5),label='RT_user_Norm')
sns.kdeplot(data=df_norm,x='Met_Norm',fill=True,clip=(0,5),label='Met_Norm')
sns.kdeplot(data=df_norm,x='Met_user_Norm',fill=True,clip=(0,5),label='Met_user_Norm')
sns.kdeplot(data=df_norm,x='IMDB_Norm',fill=True,clip=(0,5),label='IMDB_Norm')

plt.legend(loc=(1.05,0.5))
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.kdeplot(data=df_norm,x='RT_Norm',fill=True,clip=(0,5),label='RT_Norm')
sns.kdeplot(data=df_norm,x='STARS',fill=True,clip=(0,5),label='Stars')
plt.legend(loc=(1.05,0.5))
```
